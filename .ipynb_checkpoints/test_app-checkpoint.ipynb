```python
#UI implementation
import tkinter as tk
from tkinter import ttk
%store -r result_tomorrow_rain
%store -r result_max_temp
%store -r result_min_temp

root = tk.Tk()

#Set window attributes
root.title("Weather Prediction")
root.geometry("750x500")

#top_bar = tk.Label(root, text="Weather Prediction", font=('Century 40 bold')).grid(column=0, row=0, sticky=tk.W)
content = ttk.PanedWindow(orient=tk.HORIZONTAL)

left_side = Frame(content)
right_side = Frame(content)
left_side.pack(side=LEFT)
right_side.pack(side=RIGHT)

left_side.columnconfigure(0, weight=1)
right_side.columnconfigure(0, weight=1)

#Function to get inputs
def get_input(result_sheet):
    entry= ttk.Entry(result_sheet, width=40)
    entry.focus_set()
    entry.pack()
    
#Right side
style_result_sheet = ttk.Style()
style_result_sheet.configure('result_sheet.TFrame',font=('calibri',50),background='gray',highlightbackground="black",highlightthickness=2,width=150)

style_btn_submit = ttk.Style()
style_btn_submit.configure('btn_submit.TButton', font=('calibri', 12, 'bold'), foreground='blue',height=100, width=10)

#Functions to display results
def display_result(result_sheet,result,value):
    final = "Posibility for "+result+" "+value
    result = ttk.Label(result_sheet, text=final, font=('Century 20 bold'))
    result.grid(column=0, row=0, sticky=tk.W)
    
def display_RainTomorrow():
    result_sheet = ttk.Frame(right_side, style='result_sheet.TFrame')
    result_sheet.pack(pady=20,padx=20)

    attr_arr = ['Location','MinTemp','MaxTemp','Rainfall','Evaporation','Sunshine','WindGustDir','WindGustSpeed']
                #'WindDir9am','WindDir3pm','WindSpeed3pm','Humidity9am','Humidity3pm','Pressure9am','Pressure3pm',
                #'Cloud9am','Cloud3pm','Temp9am','Temp3pm','RainToday']
    input_arr = []
    for i in attr_arr:
        #e1 = tk.Entry(result_sheet)
        input_label = ttk.Label(result_sheet, text=i, width=40)
        input_label.pack()
        #get_input(result_sheet)
        #global entry
        #input_arr[count]= entry.get()
        #label.configure(text=string)
        input_arr.append(get_input(result_sheet))
    btn_submit = ttk.Button(result_sheet, text="Enter", style='btn_submit.TButton', command=display_result(result_sheet,"Rain Tomorrow",result_tomorrow_rain))
    btn_submit.pack()
    
def display_MinTemp():
    result_sheet = ttk.Frame(right_side, style='result_sheet.TFrame')
    result_sheet.pack(pady=20,padx=20)
    display_result(result_sheet,"Minimum Temparature",result_min_temp)
    
def display_MaxTemp():
    result_sheet = ttk.Frame(right_side, style='result_sheet.TFrame')
    result_sheet.pack(pady=20,padx=20)
    display_result(result_sheet,"Maximum Temparature",result_max_temp)
    
def display_EveningWeather():
    result_sheet = ttk.Frame(right_side, style='result_sheet.TFrame')
    result_sheet.pack(pady=20,padx=20)
    display_result(result_sheet,"Evening Weather",result_tomorrow_rain)
    
def display_Visualization():
    result_sheet = ttk.Frame(right_side, style='result_sheet.TFrame')
    result_sheet.pack(pady=20,padx=20)
    display_result(result_sheet,"Data Visualization",)
    
#Left side
style_Select = ttk.Style()
style_Select.configure('Select.TButton', font=('calibri', 12, 'bold'), foreground='blue',height=100, width=20)

#Select result category
btn_RainTomorrow = ttk.Button(left_side, text="Rain Tomorrow", style='Select.TButton', command=display_RainTomorrow)
btn_RainTomorrow.grid(column=0, row=0, pady=5)
btn_MinTemp = ttk.Button(left_side, text="Minimum Temparature", style='Select.TButton', command=display_MinTemp)
btn_MinTemp.grid(column=0, row=1, pady=5)
btn_MaxTemp = ttk.Button(left_side, text="Maximum Temparature", style='Select.TButton', command=display_MaxTemp)
btn_MaxTemp.grid(column=0, row=2, pady=5)
btn_EveningWeather = ttk.Button(left_side, text="Evening Weather", style='Select.TButton', command=display_EveningWeather)
btn_EveningWeather.grid(column=0, row=3, pady=5)
btn_Visualization = ttk.Button(left_side, text="Data Visualization", style='Select.TButton', command=display_Visualization)
btn_Visualization.grid(column=0, row=4, pady=5)


#place the panedwindow on the root window
content.pack(fill=tk.BOTH, expand=True)

root.mainloop()
```
